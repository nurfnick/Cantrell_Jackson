```
import pandas as pa
import numpy as np
import plotly.express as px
```

```
url = 'https://raw.githubusercontent.com/nurfnick/Cantrell_Jackson/master/Support-Oct18-Sept19.csv'
```

```
df2 = pa.read_csv(url)
```

```
df2
```

```
df2[df2['HS Customer']==18] #only ids that are 18
```

```
source = ['Email','Portal','Phone']
values = []
for s in range(len(source)):
    print('source ', source[s], ' = ', len(df2[df2["Source"]== source[s]]))
    values.append(len(df2[df2["Source"]== source[s]]))

fig = px.bar(x=source, y=values, labels={'x':'Source Type', 'y':'Total per Source Type'})

fig.show()
```
