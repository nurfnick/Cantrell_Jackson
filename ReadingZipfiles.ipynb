```
import pandas as pa
import numpy as np
from io import BytesIO
from zipfile import ZipFile
from urllib.request import urlopen
```

```
#to copy url, right click on "download" and press "copy link address"
url = urlopen('https://github.com/nurfnick/Cantrell_Jackson/raw/master/hs_customer_18.zip')
```

```
zipfile = ZipFile(BytesIO(url.read()))
df = pa.read_csv(zipfile.open('hs_customer_18.csv'))
df
```
